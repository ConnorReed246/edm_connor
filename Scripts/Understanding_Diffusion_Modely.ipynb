```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0.05  # Drift coefficient
sigma = 0.1  # Diffusion coefficient
X0 = 1.0  # Initial value
T = 1.0  # Total time
N = 100  # Number of time steps
dt = T / N  # Time step size

# Time array
t = np.linspace(0, T, N+1)

# Initialize state variable array
X = np.zeros(N+1)
X[0] = X0

# Simulate the SDE
for i in range(1, N+1):
    dW = np.random.normal(0, np.sqrt(dt))
    X[i] = X[i-1] + mu * X[i-1] * dt + sigma * X[i-1] * dW

# Plot the simulated process
plt.plot(t, X)
plt.xlabel('Time')
plt.ylabel('X(t)')
plt.title('Simulated Diffusion Process with Drift')
plt.show()
```
